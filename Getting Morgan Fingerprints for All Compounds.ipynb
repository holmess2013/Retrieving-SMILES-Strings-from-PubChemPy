```python
df = pd.read_csv('dataset.csv', sep = '\t')
```

```python
smiles_list = df['SMILES'].tolist()
names_list = df['Chemical'].tolist()
df['Mean IC50 (M)']
```

```python
# Prepare the following:

# 1) A dictionary of all morgan fingerprint RDKit objects, where the key will be the chemical name.

# 2) A dictionary containing the bit info for all fingerprints, where the key will be the chemical name. 
    
# 3) A dictionary of all RDKit molecule objects, where the key will be the chemical name.

# For each of these, you will need to add some code so that when you get to multiple compounds  with the same name, 
# you add  a number to their name so you don't try to add keys with the same name to a dictionary, which is impossible. 

    
    
    
```

```python
# For 1:
all_fp = {}
counter = 0
for name, smiles in zip(names_list, smiles_list):
    mol = AllChem.MolFromSmiles(smiles)
    fp = AllChem.GetMorganFingerprintAsBitVect(mol, 3, nBits = 1024)
    if name in all_fp:
        counter = counter + 1
        all_fp[name + "_" + str(counter)] = fp
    else:
        all_fp[name] = fp
        counter = 0
        
```

```python
# For 2:
all_bit_infos = {}
counter = 0
for name, smiles in zip(names_list,smiles_list):
    info = {}
    mol = AllChem.MolFromSmiles(smiles)
    fp = AllChem.GetMorganFingerprintAsBitVect(mol, 3, nBits = 1024, bitInfo = info)
    if name in all_bit_infos:
        counter = counter + 1
        all_bit_infos[name + '_' + str(counter)] = info
    else:
        all_bit_infos[name] = info
        counter = 0
    
```

```python
# For 3:

all_molecules = {}
counter = 0
for name, smiles in zip(names_list, smiles_list):
    mol = AllChem.MolFromSmiles(smiles)
    if name in all_molecules:
        counter = counter + 1
        all_molecules[name + '_' + str(counter)] = mol
    else:
        all_molecules[name] = mol
        counter = 0
    
    
```

```python
# Looks good! Try drawing the morgan bits for Dopamine!
prints = ([(all_molecules['Dopamine'], x, all_bit_infos['Dopamine']) for x in all_fp['Dopamine'].GetOnBits()])
Draw.DrawMorganBits(prints, molsPerRow = 5, legends = [str(x) for x in all_fp['Dopamine'].GetOnBits()])
```

```python
# You might be asking yourself, if a 1 shows up at the same index between two fingerprints for two different molecules,
# is it the same substructure? The answer is yes, but let's show that by drawing the substructures for two very similar
# molecules from this dataset: 4-Nonylphenol and 4-Octylphenol.
```

```python
# Substructures for 4-Nonylphenol
prints = ([(all_molecules['4-Nonylphenol'], x, all_bit_infos['4-Nonylphenol']) for x in all_fp['4-Nonylphenol'].GetOnBits()])
Draw.DrawMorganBits(prints, molsPerRow = 5, legends = [str(x) for x in all_fp['4-Nonylphenol'].GetOnBits()])
```

```python
# Substructures for 4-Octylphenol
prints = ([(all_molecules['4-Octylphenol'], x, all_bit_infos['4-Octylphenol']) for x in all_fp['4-Octylphenol'].GetOnBits()])
Draw.DrawMorganBits(prints, molsPerRow = 5, legends = [str(x) for x in all_fp['4-Octylphenol'].GetOnBits()])
```

```python
# Finally, let's make a CSV where we have the chemical name, the morgan fingerprint, and the IC50 in the dataframe. 
```

```python
all_fp_list = []
for fp in all_fp.values():
    fp_array = np.array(fp)
    fp_list = fp_array.tolist()
    all_fp_list.append(fp_list)
```

```python
new_df = pd.DataFrame()
```

```python
new_df['Chemical'] = all_fp.keys()
new_df['Morgan Fingerprint'] = all_fp_list
IC50_list = df['Mean IC50 (M)'].tolist()
new_df['Mean IC50 (M)'] = IC50_list
```
